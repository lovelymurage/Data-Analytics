```python
#11. ANOVA test
import scipy as scipy
from scipy import stats as st
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#i. Whether work experience influences the salary amount for the employees.
#null hypothesis (HO): Work Experience has no significant effect on salary being issued
#alternative hypothesis (H1): Work Experience has a significant effect on the salary being issued
one_way_anova = ols('Monthly_Salary ~ Work_Experience', data = saldata).fit()
one_way = sm.stats.anova_lm(one_way_anova, type = 2 )
one_way
```

```python
#p-value is 0.507272 which is greater than 0.05 there is no significant evidence to reject the null hypothesis
#therefore Work Experience has no significant effect on the salary being issued 
```

```python
#ii. Whether education level influences the salary amount for the employees.
#null hypothesis (HO): Educational level has no significant effect on salary being issued
#alternative hypothesis (H1): Educational level has a significant effect on the salary being issued
one_way_anova = ols('Monthly_Salary ~ Education_Level', data = saldata).fit()
one_way = sm.stats.anova_lm(one_way_anova, type = 2 )
one_way
```

```python
#p-value is 0.187077 which is greater than 0.05 there is no significant evidence to reject the null hypothesis
#therefore Education level has no significant effect on the salary being issued 
```

```python
#iii. The interaction effect for Education Level and Work Experience on the salary.
#null hypothesis (HO): Educational level and Work experience have no significant effect on salary being issued
#alternative hypothesis (H1): Educational level and Work Experience have a significant effect on the salary being issued
one_way_anova = ols('Monthly_Salary ~ Work_Experience*Education_Level', data = saldata).fit()
one_way = sm.stats.anova_lm(one_way_anova, type = 2)
one_way
```

```python
#p-value for both variables combined is 0.873539 > 0.05
#therefore the test isn't significant
#conclusion: there is no sufficient evidence to reject the null hypothesis
#hence the organization doesn't prioritize both work experience and educational level when awarding salaries
```

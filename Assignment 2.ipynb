```python
#4. Chi-Square Test of Independence
#a. Test whether there is a significant association between Gender and Smoking_Status
import scipy as scipy
from scipy import stats as st
from scipy.stats import chi2_contingency
#Create a contingency table to show the observed frequencies for each combination of categories
contingency_table1 = pd.crosstab(healthdata['Gender'], healthdata['Smoking_Status'])
print("Contingency Table (Observed Frequencies):")
print(contingency_table1)
```

```python
#H0: There is no significant association between Gender and Smoking_Status
#H1: There is a significant association between Gender and Smoking_Status 
chi2, p_value, dof, expected_frequencies = chi2_contingency(contingency_table1)
print(f"\nChi-square Statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:")
print(expected_frequencies)
```

```python
#p-value = 1.0 which is >0.05 thus there is no evidence to reject the null hypothesis
#conclusion: there is no significant association between Gender and Smoking_Status
```

```python
#b. Test whether there is a significant association between Diabetes and Blood_Pressure
#Create a contingency table to show the observed frequencies for each combination of categories
contingency_table2 = pd.crosstab(healthdata['Diabetes'], healthdata['Blood_Pressure'])
print("Contingency Table (Observed Frequencies):")
print(contingency_table2)
```

```python
#H0: There is no significant association between Diabetes and Blood_Pressure
#H1: There is a significant association between Diabetes and Blood_Pressure
chi2, p_value, dof, expected_frequencies = chi2_contingency(contingency_table2)
print(f"\nChi-square Statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:")
print(expected_frequencies)
```

```python
#p-value = 0.9963721493180326 which is >0.05 thus there is no evidence to reject the null hypothesis
#conclusion: there is no significant association between Diabetes and Blood_Pressure
```

```python
#5. T-Test (Independent Samples): Test if the mean BMI differs significantly between Male and Female patients.
import scipy as scipy
from scipy import stats as st
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#H0: The mean BMI doesn't differ significantly between Male and Female patients.
#H1: The mean BMI differs significantly between Male and Female patients.
# Separate the BMI data into two independent groups (Male and Female)
BMI_Male = healthdata[healthdata['Gender'] == 'Male']['BMI']
BMI_Female = healthdata[healthdata['Gender'] == 'Female']['BMI']
#Performing the Independent T-test
t_statistic, p_value = st.ttest_ind(BMI_Male, BMI_Female, equal_var=False)
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

```python
#p-value = 0.8269 which is > 0.05, there is no significant evidence to reject the null hypothesis
#conclusion: the mean BMI doesn't differ significantly between male and female patients
```

```python
#finding the mean BMI of Male patients
np.mean(BMI_Male)
```

```python
#finding the mean BMI of Female patients
np.mean(BMI_Female)
```

```python
#6. One-Way ANOVA
# Test if the mean Exercise_Frequency differs significantly across the different Medication_Adherence levels (Poor, Average, Good).
healthdata.Medication_Adherence.value_counts()
```

```python
#HO: The mean Exercise_Frequency doesn't differ significantly across the different Medication_Adherence levels (Poor, Average, Good)
#H1: The mean Exercise_Frequency differs significantly across the different Medication_Adherence levels (Poor, Average, Good)
```

```python
one_way_anova = ols('Exercise_Frequency ~ C(Medication_Adherence) ', data = healthdata).fit()
one_way = sm.stats.anova_lm(one_way_anova, type = 2)
one_way
```

```python
#p-value = 0.95774 is > 0.05, there is no significant evidence to reject the null hypothesis.
#Conclusion: The mean Exercise_Frequency doesn't differ significantly across the different Medication_Adherence levels (Poor, Average, Good)
```

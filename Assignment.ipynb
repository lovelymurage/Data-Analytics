#### Filling the missing values
- The 8 missing values in the stud_id column can't be filled since they are a unique identifier

#### 1. Using the measures of central tendency
a) Filling missing values in final_exam variable using the mean since preserves the overall average of the dataset.

```python
df.final_exam = df.final_exam.fillna(df.final_exam.mean())
```

```python
#checking if the missing values in the final_exam variable have been filled
df.final_exam.isnull().sum()
```

b) Filling missing values in fam_income_kes variable using median because median represents the "typical" family income without being pulled by extremely high or low values, i.e. outliers since income data is typically slewed.

```python
df.fam_income_kes = df.fam_income_kes.fillna(df.final_exam.mean())
```

```python
#checking if the missing values in the fam_income_kes variable have been filled
df.fam_income_kes.isnull().sum()
```

c) Filling missing values in lecturer_notes variable using mode (the most occuring value) because it is a categorical variable.

```python
#the mode of the variable lecturer_notes
df.lecturer_notes.mode()
```

```python
#Using the mode to fill the missing values of the variable lecturer_notes
df.lecturer_notes = df.lecturer_notes.fillna('Good')
```

```python
#checking if the missing values in the lecturer_notes variable have been filled
df.lecturer_notes.isnull().sum()
```

#### 2. Using the Forward Fill (ffill) method to fill missing values in age_yrs variable because students with nearby IDs are likely have similar ages.

```python
df.age_yrs = df.age_yrs.ffill()
```

```python
#checking if the missing values in the lecturer_notes variable have been filled
df.age_yrs.isnull().sum()
```

#### 3. Using the Backward Fill (bfill) method to fill missing values in assgn_score because It's a good placeholder as it shows an actual score from the dataset.

```python
df.assgn_score = df.assgn_score.ffill()
```

```python
#checking if the missing values in the assgn_score variable have been filled
df.assgn_score.isnull().sum()
```

#### Difference between forward fill and backward fill methods
- The difference is in the direction in which known valid observations are propagated to fill in missing data.
- Forward fill method fills in missing values by propagating the last known valid observation forward to the next missing one.
- Backward fill method fills in missing values by using the next known valid observation backward to the previous missing one.

### Standardization

```python
#Categories in the sex variavle
df.sex.value_counts()
```

- There are inconsistencies in the categories since both the male and female students are identified using different variations instead of just one identification, i.e, for male students there is Male, male, M and for female students there is Female, female and F.

```python
#Standardizing the format of the sex variable
df.sex = df.sex.replace({'F':'Female', 'female': 'Female', 'male':'Male', 'M': 'Male'})
```

```python
#verifying the standardization was successful
df.sex.value_counts()
```

#### Importance of data cleaning and preprocessing before conducting statistical analysis
- Improves data accuracy: Removes errors and duplicate records that can distort statistical measures like means, variances, and correlations.
- Handles Missing Information: Prevents misleading interpretations since missing values can significantly affect analysis outcomes.
- Reduces noise and inconsistency to improve the detection of meaningful trends and relationships.
- Enhancing data integration: when combining data from multiple sources, preprocessing helps create a unified and coherent dataset.
